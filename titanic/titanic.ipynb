```python
# data analysis and wrangling
import pandas as pd
import numpy as np
import random as rnd

# visualization
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Retirando dados nulos da coluna age
#train_df = train_df[np.isfinite(train_df['Age'])]
train_df.dropna(subset=['Age'])

#Funcao responsavel por remover as colunas passadas por parâmetro
def remove(colunas):
    for i in range(0, len(colunas)):
        train_df.drop([colunas[i]],axis = 1, inplace = True)
    return train_df.head()
#Colunas a serem removidas
colunas = ['Pclass','Fare','Ticket','Cabin','PassengerId','SibSp']
#Chama a funcao e passa o parâmetro das colunas a serem removidas
remove(colunas)


```

```python
print('Sobreviventes     :', train_df['Survived'].value_counts()[1])
print('Não Sobreviventes :', train_df['Survived'].value_counts()[0])

#Gera o grafico
train_df['Survived'].value_counts().plot.pie(colors=('tab:red', 'tab:blue'),title='Qual a porcentagem dos passageiros sobreviventes?',fontsize=13, shadow=True,startangle=90,autopct='%1.1f%%',
labels=('Não sobreviventes','Sobreviventes'),
figsize=(5,5)).set_ylabel('')
```

```python
#Gera histograma de frequencia de idade
#print(type(train_df))
plt.figure();
train_df.hist(column='Age', color=('blue'), alpha=0.8, bins=10, figsize=(10,4))
plt.title('Faixa Etária dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequência')
```

```python
#Divide os dados em grupos de criancas e adultos
train_df['Crianca_Adulto'] = 0
train_df.loc[train_df[train_df['Age'] < 18].index, 'Crianca_Adulto'] = 'Crianças'
train_df.loc[train_df[train_df['Age'] >= 18].index, 'Crianca_Adulto'] = 'Adulto'

#train_df[:30]


#Gera grafico de sobrevivencia de criancas e adultos
train_df.groupby('Crianca_Adulto')['Survived'].mean().plot.barh(title='4.3. Média de Sobrevivência', figsize=(10,2.5), color=('tab:green','tab:blue')).set_ylabel(''),plt.xlabel('')
```
